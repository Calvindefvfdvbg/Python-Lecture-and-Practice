```python
# Date and time in python


import datetime


# Get the current date and time
now = datetime.datetime.now()


print(now)
```

```python
# get only month
print(now.month)

# get only day
print(now.day)

# get only year
print(now.year)



# get only hour
print(now.hour)

# get only minute
print(now.minute)

# get only second
print(now.second)
```

```python
# get complete date 
print(now.strftime("%Y-%m-%d"))

# get complete time
print(now.strftime("%H:%M:%S"))


# get month name in character
print(now.strftime("%B"))
```

```python
# List in python
# List is a collection which is ordered and changeable. Allows duplicate members
```

```python
# create a empty list
my_list = []
print(my_list)      # Output: []
# check the type of my_list
print(type(my_list))     # Output: <class'list'>
```

```python
# init list with values
my_list = [1,2,3,4,5]
print(my_list)      #Output: [1,2,3,4
```

```python
# mixed data types
my_list = [1,2,3,4,5,"hello",3.14]
print(my_list)     #Output: [1,2,3,4,5,"hello",3.14]
```

```python
# aecess list element
#index       0        1          2        3        4        5       6
fruits = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
#-ve         -7       -6         -5       -4       -3       -2      -1


# print the element at index 0
print(fruits[0])         #Output: apple
print(fruits[-7])         #Output: apple


print(fruits[1])         #Output: banana
```

```python
#slice list
# get the elements from index 1 to 4 (not included)
print(fruits[1:4])       #Output: ['banana','cherry','orange']
print(fruits[-6:-3])     #Output: ['banana','cherry','orange']

print(fruits[:4])        #Output: ['appple','banana','cherry','orange']

print(fruits[:2])

print(fruits[:])
print(fruits)
```

```python
print(fruits)
# change the value of an element
fruits[0] = "avocado"
print(fruits)
```

```python
# insert element at specific index
fruits.insert(2,"blueberry")
print(fruits)
```

```python
# append element at the end of the list
fruits.append("strawberry")
print(fruits)
```

```python
# remove element
fruits.remove("banana")
print(fruits)
```

```python
# extend list
new_fruits = ["grapes","watermelon"]
fruits.extend(new_fruits)
print(fruits)
```

```python
# pop element from the end of the list
fruits.pop()
print(fruits)
```

```python
# intreting list
for i in fruits:
    print(i)
```

```python
# check if n element is in the list


print("apple" in fruits)

print("kiwi" in fruits)
```

```python
# how to take user input into list


# define size and list


size = int(input("Enter the size of the list:"))
food = []


# take input from user
for i in range(size):
    f = input("Enter the food:")
    food.append(f)

for i in food:
    print(i)
```
